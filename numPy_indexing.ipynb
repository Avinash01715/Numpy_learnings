Indexing and slicing in matrix

```python
arr1 = np.arange(0,30).reshape(5,6)
arr1
```

```python
arr1[0:2,1:3]
```

```python
slice = arr1[2: ,4: ]
slice
```

```python
slice_col = arr1[: , 3]
slice_col
```

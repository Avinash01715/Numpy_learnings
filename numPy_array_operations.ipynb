## DEEP AND SHALLOW COPY

```python
a = np.arange(1,21)
a
```

```python
b = a + 10
b
```

```python
a   #shallow copy no change in original
```

```python
c = a 
c
```

```python
c[0] = 23
c
```

```python
a #deep copy original is changed that also happens in list
```

```python
d = a.copy()     # can be made shallow copy 
d
```

```python
d[0]= 34
d
```

```python
a
```

# Advanced array manipulations

> Stacking Array

```python
x = np.array([2,32,1,3])
y = np.array([7,36,3,7])
```

```python
np.hstack((x,y))
```

```python
np.vstack((x,y))
```

```python
np.column_stack((x,y))
```

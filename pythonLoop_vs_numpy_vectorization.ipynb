```python
import numpy as np
import time

np.random.seed(42)
rows = 100000
cols = 50

data = np.random.randn(rows,cols).astype(np.float32)
weight = np.random.randn(cols).astype(np.float32) * 10

```

```python
# loop version

def loop_weighted_sum(data,weight):
    result = np.zeros(rows, dtype = np.float32)
    for i in range(rows):
        for j in range(cols):
            result[i] += data[i,j] * weight[j]   
    return result


start_time = time.time()
res = loop_weighted_sum(data, weight)
loop_time = time.time() - start_time

print(f"Loop took {loop_time:.4f} seconds")



            
```

```python
#Numpy vectorized version

def vectorized_weighted_sum(data,weight):
    vec_result = np.sum(data*weight, axis = 1)
    return vec_result

start_time = time.time()
res = vectorized_weighted_sum(data, weight)
vectorized_time = time.time() - start_time

print(f"Vectorization took {vectorized_time:.4f} seconds")

    


```

```python
speedup = loop_time / vectorized_time
print(f"{speedup:.2f}")
```
